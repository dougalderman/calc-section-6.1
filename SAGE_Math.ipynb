```python
v = var('v')
```

```python
T = var('T')
```

```python
def W(v,T): return 91.4 - ((10.45 + 6.68 * sqrt(v) - 0.447*v)*(457 - 5*T)) / 110
```

```python
P = plot3d(W,(v,0,50),(T,0,100),color='blue')
```

```python
P.show()
```

```python
Q = plot3d(W,(v,0,50),(T,0,100),adaptive=True,color='blue')
```

```python
Q.show()
```

```python
R = plot3d(W,(v,0,50),(T,0,100),color='blue',opacity=0.8)
```

```python
R.show()
```

```python
P1 = plot3d(W,(v,0,50),(T,0,100),frame=False,color='blue')
```

```python
P1.show()
```

```python
P2 = plot3d(W,(v,0,50),(T,0,100),adaptive=True,color=rainbow(60, 'rgbtuple'))
```

```python
P2.show()
```
